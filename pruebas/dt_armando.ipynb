```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = {"Cant. Trabajadores": ["S.O",1,2,3,4,5],
     "Espacial": [0, 67, 74, 78, 83, 89],
     "Temporal": [0, 75, 79, 84, 88, 94],
     "Fusion":[125.3, 125.37, 76.7,  50.78, 38.34, 26.59],
     "Media-Clase":[115.55, 115.51, 73.34, 52.644, 31.87, 19.43],
     "Correlacion":[110.3, 110.35, 66.48, 44.21, 33.74, 23.02],
     "Regresion Lineal":[133.41, 133.6, 78.8, 56.71, 41.46, 34.72]}
df_data = pd.DataFrame(data=df)
df_data["Total"] =  df_data.sum(axis=1)
```

```python
nameColumn = "Total"
times="Segundos"
fig, ax = plt.subplots(figsize=(10, 10))
sns.set(font_scale = 1.7)
sns.barplot(x="Cant. Trabajadores",y=nameColumn, data=df_data, linewidth=2, saturation=4)
ax.bar_label(ax.containers[0])
plt.ylabel("Tiempos de servicio - {} \n {}".format(nameColumn,times))
plt.title("Tiempos de servicio - {}".format(nameColumn), fontsize = 20)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.set(font_scale = 1.7)
sns.barplot(x="Cant. Trabajadores",y="Tiempos de Servicio - Fusion", data=df_data, linewidth=2, saturation=4)
ax.bar_label(ax.containers[0])
plt.ylabel("Tiempos de servicio - Fusion \n Segundos")
plt.title("Tiempos de servicio - Temporal", fontsize = 20)
```

```python
df_2019 = {"Test": ["All file","Only date"],
     "Read Time": [131.67, 25.5],
     "Conteo Time":[6.6,1.3]}
df_2019 = pd.DataFrame(data=df_2019)
df_2019
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(y="Read Time",x="Test", data=df_2019, linewidth=2, saturation=4)
plt.title("READ TIME - 2019", fontsize = 20)
plt.ylabel("Time (s)")
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(y="Conteo Time",x="Test", data=df_2019, linewidth=2, saturation=4)
plt.title("COUNT TIME - 2019", fontsize = 20)
plt.ylabel("Time (s)")
```

```python
df_2019["Total"] = df_2019["Conteo Time"] + df_2019["Read Time"]
df_2019
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(y="Total",x="Test", data=df_2019, linewidth=2, saturation=4)
plt.title("TOTAL TIME - 2019", fontsize = 20)
plt.ylabel("Time (s)")
```
